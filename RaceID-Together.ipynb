```r
# validate
drt %>% filter(name=="FO538757.2")
```

```r
drt %>% 
   mutate(ctr.perc = 100 * ctr / sum(ctr)) %>% 
   mutate(sin.perc = 100 * sin / sum(sin)) %>% 
   arrange(desc(ctr.perc))
```

What is the proportion of mitochondrial genes?

```r
data.frame(
    ctr=100 * sum(mt.expr$ctr) / sum(drt$ctr),
    sin=100 * sum(mt.expr$sin) / sum(drt$sin))
```

13 MT genes, all with very similar levels of expression between CTR and SIN (i.e. not at all DE), and with less than 4% contributing to the variance in the data.

#### Centred distributions

Before we perform clustering on the above set, let's first clip off the under and over expressed cells on our gene-filtered set.

```r
options(repr.plot.height = 5)
par(mfrow=c(2,2))
hist(sce.ctr$total_counts, breaks=100, main="LibSize CTR (¬MT)")
abline(v=c(10000,60000), col='red')
hist(sce.sin$total_counts, breaks=100, main="LibSize SIN (¬MT)")
abline(v=c(10000,60000), col='red')
hist(sce.ctr.no_rp$total_counts, breaks=100, main="LibSize CTR (¬MT¬RP)")
abline(v=c(5000,30000), col='blue')
hist(sce.sin.no_rp$total_counts, breaks=100, main="LibSize SIN (¬MT¬RP)")
abline(v=c(8000,30000), col='blue')
```

When we exlude only MT-genes the distributions can be centred if we keep the 10k-60k range
When we exlude the MT and RP-genes, the distributions centre more around the 5k-30k range.

Regardless of how they are distributed, let us clip off the > 30K and <5K in order to see if our over expressed transcripts are still over expressed.

#### RAW ¬MT : 10-60K

```r
sce.sin.cfilt <- sce.sin[,(sce.sin$total_counts > 10000) & (sce.sin$total_counts < 60000)]
sce.ctr.cfilt <- sce.ctr[,(sce.ctr$total_counts > 10000) & (sce.ctr$total_counts < 60000)]
```

```r
data.frame(sin=dim(sce.sin), sinfilt=dim(sce.sin.cfilt), ctr=dim(sce.ctr), ctrfilt=dim(sce.ctr.cfilt), row.names = c("Genes", "Cells"))
```

We lose only 200-300 cells, this is fine.

```r
sce.sin.cfilt <- calculateQCMetrics(sce.sin.cfilt)
sce.ctr.cfilt <- calculateQCMetrics(sce.ctr.cfilt)
```

```r
plotHighestExprs(sce.sin.cfilt)
plotHighestExprs(sce.ctr.cfilt)
```

We see the same expression of FTH1 of 8% and 2% shown in the original raw (sans MT) matrices.

Let's repeat this for the sans RP matrices too using the 5K-30K range.

#### RAW ¬MT¬RP : 5-30K

```r
sce.sin.no_rp.cfilt <- sce.sin.no_rp[,(sce.sin.no_rp$total_counts > 5000) & (sce.sin.no_rp$total_counts < 30000)]
sce.ctr.no_rp.cfilt <- sce.ctr.no_rp[,(sce.ctr.no_rp$total_counts > 5000) & (sce.ctr.no_rp$total_counts < 30000)]

data.frame(
    sin=dim(sce.sin), sinfilt=dim(sce.sin.cfilt), sinrpfilt=dim(sce.sin.no_rp.cfilt),
    ctr=dim(sce.ctr), ctrfilt=dim(sce.ctr.cfilt), ctrrpfilt=dim(sce.ctr.no_rp.cfilt),
    row.names = c("Genes", "Cells")
)
```

We see a loss of ~300 in both, again fine.

```r
sce.sin.no_rp.cfilt <- calculateQCMetrics(sce.sin.no_rp.cfilt)
sce.ctr.no_rp.cfilt <- calculateQCMetrics(sce.ctr.no_rp.cfilt)

plotHighestExprs(sce.sin.no_rp.cfilt)
plotHighestExprs(sce.ctr.no_rp.cfilt)
```

When we compare this to the RP matrices ("#### Without Ribosome coding genes"), we see that the FTH1 has a more centred distribution in the mutant (though still high) but retains a lower-heavy distribution for the control.

Interestingly this behaviour persists, even if we set the cell filtering range on the RP matrices to between 10k and 20k. This means that the expression of FTH1 does not originate from oversaturated cells or undersaturated ones, but is present in a significant portion of our "good" cells.

#### Summary

In the previous section we explored N setups:

 1. *RAW*
    * We observed that our most DE genes were RPL RPS and FTH1
 1. *RAW ¬MT*
    * We observed MT made a very small contribution to the variance, but we filtered them out. FTH1 was 2% in the control (skewed left) and 8% in the SIN (centred)
 1. *RAW ¬MT¬RP*
    * We filtered out any ribosomal genes and mitochondrial genes, and saw  that FTH1 was still highly expressed, now with 4% in the control (mean towards the lower-tail of distribution), and the mutant steady at 8% (centred distrib.)
    
We then looked at the library sizes of the cells to see if we could set sensible filtering thresholds that would center the count distributions and hopefully reduce the contributing variance from overexpressed cells that might be overexpressing FTH1.

We discovered 10-60K would fit the *RAW¬MT* distributions, and 5-30K would fit *RAW¬MT¬RP* distributions
    
 1. RAW ¬MT:10-60K
    * We saw the expression FTH1 at 2% in the control which did not change when compared against the *RAW ¬MT*, and still retain it's skewed distribution. Over expressed cell removal did nothing.
    
 1. RAW ¬MT¬RP:5-30K
    * We saw the expression of FTH1 remain at 4% in the control, but jumped to 15% in the mutant(!!) 
    
We will now try the same but removing FTH1 from the analysis

```r
plotHighestExprs(sce.sin.no_rp_fth1)
plotHighestExprs(sce.ctr.no_rp_fth1)
```

* MALAT1 appears to be over-expressed in the mutant, but normally expressed in the control.
* EEF1A1 appears to be over-expressed in both, but a 2.5% in control and 4% in mutant (i.e. DE)

#### RAW -MT-RP : 5-30K

```r
options(repr.plot.height = 5)
par(mfrow=c(2,2))
hist(sce.ctr.no_rp$total_counts, breaks=100, main="LibSize CTR (¬MT¬RP)")
abline(v=c(5000,30000), col='blue')
hist(sce.ctr.no_rp_fth1$total_counts, breaks=100, main="LibSize CTR (¬MT¬RP¬FTH1)")
abline(v=c(5000,30000), col='blue')
hist(sce.sin.no_rp$total_counts, breaks=100, main="LibSize SIN (¬MT¬RP)")
abline(v=c(5000,30000), col='red')
hist(sce.sin.no_rp_fth1$total_counts, breaks=100, main="LibSize SIN (¬MT¬RP¬FTH1)")
abline(v=c(5000,30000), col='red')
```

Removal of FTH1 appears to remove some of the overexpressed cells in both distributions. 5k-30k still seems a sensible threshold of filtering.

```r
sce.ctr.no_rp_fth1.cfilt <- sce.ctr.no_rp_fth1[,(sce.ctr.no_rp_fth1$total_counts > 5000) & (sce.ctr.no_rp_fth1$total_counts < 30000)]
sce.sin.no_rp_fth1.cfilt <- sce.sin.no_rp_fth1[,(sce.sin.no_rp_fth1$total_counts > 5000) & (sce.sin.no_rp_fth1$total_counts < 30000)]

data.frame(
    sin=dim(sce.sin), sinfilt=dim(sce.sin.no_rp_fth1.cfilt),
    ctr=dim(sce.ctr), ctrfilt=dim(sce.ctr.no_rp_fth1.cfilt),
    row.names = c("Genes", "Cells")
)
```

```r
sce.ctr.no_rp_fth1.cfilt <- calculateQCMetrics(sce.ctr.no_rp_fth1.cfilt)
sce.sin.no_rp_fth1.cfilt <- calculateQCMetrics(sce.sin.no_rp_fth1.cfilt)
```

```r
plotHighestExprs(sce.sin.no_rp_fth1)
plotHighestExprs(sce.ctr.no_rp_fth1)
```

Barely any change to the previous RAW -MT-RP-FTH1 when using thresholds, but still good to remove background excess

```r
tmp <- list(
    ctr.mt.rp = sce.ctr.no_rp,
    ctr.mt.rp.range = sce.ctr.no_rp.cfilt,
    ctr.mt.rp.fth1 = sce.ctr.no_rp_fth1,
    ctr.mt.rp.fth1.range = sce.ctr.no_rp_fth1.cfilt,
    
    sin.mt.rp = sce.sin.no_rp,
    sin.mt.rp.range = sce.sin.no_rp.cfilt,
    sin.mt.rp.fth1 = sce.sin.no_rp_fth1,
    sin.mt.rp.fth1.range = sce.sin.no_rp_fth1.cfilt
)

saveRDS(tmp, file="checkpoints/raceid.together.all.rds")
```
