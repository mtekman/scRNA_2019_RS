There appear to be two count matrices in "MatrixMarket" format, as well two files for barcodes (cell headers) and genes (rowname features).

It is not clear whether these are two batches of the same data, but I would guess that "CTR" means control and "siN6AMT1" is the mutant N6 Methyladine modification.

## Construct full tabular matrices

We need to convert from MatrixMarket format into a full count matrix for both sets

### CTR Matrix

```r
library(Matrix)
ctr.mat <- readMM("source/CTR_matrix.mtx")
```

```r
# quick peek
dim(ctr.mat)
ctr.mat[1:10,1:10]
```

33694 rows (genes) and 3709 columns (barcodes). Let's see if we can recapitulate this in the other files for this set and merge them into one sparse matrix.

```r
ctr.bar <- read.table("source/CTR_barcodes.tsv", header=F, stringsAsFactors=F)[,1]
head(ctr.bar)
message("Barcodes:", length(ctr.bar))
```

```r
ctr.genes <- read.table("source/CTR_genes.tsv", header=F, stringsAsFactors=F)
head(ctr.genes)
message("Genes:", nrow(ctr.genes))
```

So we can indeed get the same dimensions as expected. We will use V2 names since they appear to be more informative. Let us first check that there are no duplicates

### Check for Duplicate Gene Names

```r
goodnames <- ctr.genes[,2]
dupes <- goodnames[duplicated(goodnames)]
message("Duplicate Gene names:", length(dupes))
dupes
dupes.tab <- ctr.genes[ctr.genes[,2] %in% dupes,]
dupes.tab
```

So we see 34 duplicates meaning we must use transcript names instead of gene names.

## Unite CTR matrix data and headers

One major caveat here: We must assume barcode and gene ordering given by the files.

```r
rownames(ctr.mat) <- ctr.genes[,1]
colnames(ctr.mat) <- ctr.bar
ctr.mat[1:10,1:3]
```

Let's do the same with the mutant matrix, and then checkpoint everything.

## Unite siN6AMT1 matrix data and headers

```r
sin.mat <- readMM("source/siN6amt1_matrix.mtx")
sin.bar <- read.table("source/siN6amt1_barcodes.tsv", header=F, stringsAsFactors=F)[,1]
sin.genes <- read.table("source/siN6amt1_genes.tsv", header=F, stringsAsFactors=F)[,1]

message("Mutant matrix: ", nrow(sin.mat), " x ", ncol(sin.mat))
message("Barcodes: ", length(sin.bar))
message("Genes: ", length(sin.genes))

rownames(sin.mat) <- sin.genes
colnames(sin.mat) <- sin.bar

sin.mat[1:10,1:3]
```

# Clustering with RaceID

```r
library(RaceID)
```

Let us start first with the CTR count matrix and try to see what clusters we find

```r
ctr.mat <- readRDS("checkpoints//ctrmat.rds")
```

```r
ctr.sc <- SCseq(ctr.mat)
dim((ctr.sc))
```

```r
ctr.sc <- filterdata(ctr.sc, mintotal=2000)
dim(getfdata(ctr.sc))
```

```r
ctr.sc <- compdist(ctr.sc)
```

```r
ctr.sc <- clustexp(ctr.sc)
```

```r
plotsaturation(ctr.sc, disp=F)
plotsaturation(ctr.sc, disp=T)
```

```r
plotjaccard(ctr.sc)
```

```r
ctr.sc <- findoutliers(ctr.sc)
```

```r
plotbackground(ctr.sc)
plotsensitivity(ctr.sc)
plotoutlierprobs(ctr.sc)
```

```r
clustheatmap(ctr.sc)
```

```r
ctr.sc <- comptsne(ctr.sc)
ctr.sc <- compfr(ctr.sc, knn=5)
```

```r
plotmap(ctr.sc, fr=F)
plotmap(ctr.sc, fr=T)
```

```r
saveRDS(ctr.mat, "checkpoints/ctrmat.vanilla")
```
